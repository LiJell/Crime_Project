```python
df0 = pd.DataFrame(df[0])
df1 = pd.DataFrame(df[1])
df2 = pd.DataFrame(df[2])
df3 = pd.DataFrame(df[4])
df4 = pd.DataFrame(df[5])
df5 = pd.DataFrame(df[6])
df6 = pd.DataFrame(df[7])
df7 = pd.DataFrame(df[8])
df8 = pd.DataFrame(df[9])
df9 = pd.DataFrame(df[11])
df10 = pd.DataFrame(df[13])
df11 = pd.DataFrame(df[14])
```

```python
df1.info()
df_list = [df0, df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11]
```

```python
df_total = pd.concat(df_list, ignore_index = True)
```

```python
df_total['소재지'].str.split(' ')[0][0]
```

```python
list = []
for i in range(len(df_total)):
    list.append(df_total['소재지'].str.split(' ')[i][0])

df_total['시'] = list
```

```python
df_seoul_total = df_total[df_total.시 == '서울특별시']
```

```python
df_seoul_total.sort_values('소재지', inplace = True)
```

```python
df_seoul_total.drop_duplicates(['철도역'], ignore_index = True, inplace = True)
```

```python
df_seoul_total.to_excel('../data/seoul_subway_total.xlsx', index = False)
```
