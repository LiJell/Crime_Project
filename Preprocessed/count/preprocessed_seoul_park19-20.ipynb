```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('../park/park2019.txt' , header=2, sep = '\t')
df
```

```python
df.info()
```

```python
df=df[['기간','자치구','공원수']]
df
```

```python
df['공원수']=df.loc[:,'공원수'].replace(',','')
```

```python
df['공원수'] = df.loc[:, '공원수'].str.replace(',', '').astype('int64')
```

```python
df
```

```python
df=df.drop(26)
```

```python
df.to_excel('../park/preprocessed_park2019.xlsx' , index=False)
```
