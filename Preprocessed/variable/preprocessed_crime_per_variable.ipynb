```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_excel('./crime/data_final/data_total_2019.xlsx')
```

```python
data
```

```python
data2 = data[['기간', '자치구']]
```

```python
data2['인구수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '인구 수(명)']
data2['경찰관서 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '경찰관서 수']
data2['소방관서 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '소방관서 수']
data2['카페 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '카페 수']
data2['편의점 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '편의점 수']
data2['공원 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '공원 수']
data2['버스정류장 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '버스정류장 수']
data2['가로등 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '가로등 수']
data2['ATM 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, 'ATM 수']
data2['반려동물 가구비율 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '반려동물 가구 비율(%, 인구 수)']
data2['행복지수 종합 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '행복지수 종합']
```
