```python
import pandas as pd
```

```python
#가로등 위치정보 파일 불러오기
df = pd.read_excel('./서울시가로등위치_2020.xlsx')
df
```

```python
#필요한 열만 추출
seoul_light_2020= df[['구분','가로등']]
```

```python
seoul_light_2020
```

```python
# 필요없는 행 제거
seoul_light_2020 = seoul_light_2020.iloc[9:,:]
seoul_light_2020
```

```python
# index 새로 지정
seoul_light_2020.reset_index(inplace=True)
```

```python
seoul_light_2020
```

```python
# 기존index열 제거
seoul_light_2020_cnt = seoul_light_2020.drop(['index'], axis=1)
```

```python
seoul_light_2020_cnt
```

```python
# 열이름 변경
seoul_light_2020_cnt.rename(columns={'구분':'자치구'}, inplace=True)
```

```python
# 저장
seoul_light_2020_cnt.to_excel('./seoul_light_cnt_2020.xlsx')
```
