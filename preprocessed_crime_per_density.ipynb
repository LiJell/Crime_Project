```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('../data_total_2019_pre_area.xlsx')
data2 = pd.read_excel('./crime/data_final/data_total_2019.xlsx')
```

```python
data['범죄발생건수'] = data2['범죄발생건수']
```

```python
data3 = data2[['기간', '자치구']]
```

```python
data3['인구밀도 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '인구 수(명)']
data3['경찰관서밀도 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '경찰관서 수']
data3['소방관서밀도 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '소방관서 수']
data3['카페밀도 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '카페 수']
data3['편의점밀도 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '편의점 수']
data3['공원밀도 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '공원 수']
data3['버스정류장밀도 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '버스정류장 수']
data3['가로등밀도 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, '가로등 수']
data3['ATM밀도 수 대비 범죄발생건수']= data.loc[:, '범죄발생건수']/data.loc[:, 'ATM 수']
```
