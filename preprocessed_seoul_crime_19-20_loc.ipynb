```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_table('./crime/crime_location(2014-2020).txt')
```

```python
data = data.replace('-', 0)
```

```python
data = data[-12:]
```

```python
data
```

```python
data['계'] = data['계'].str.replace(',', '').astype('int64')
data['아파트 연립다세대'] = data['아파트 연립다세대'].str.replace(',', '').astype('int64')
data['단독주택'] = data['단독주택'].str.replace(',', '').astype('int64')
data['노상'] = data['노상'].str.replace(',', '').astype('int64')
data['상점'] = data['상점'].str.replace(',', '').fillna(0).astype('int64')
data['숙박업소목욕탕'] = data['숙박업소목욕탕'].str.replace(',', '').astype('int64')
data['유흥접객업소'] = data['유흥접객업소'].str.replace(',', '').astype('int64')
data['사무실'] = data['사무실'].str.replace(',', '').astype('int64')
data['역대합실'] = data['역대합실'].str.replace(',', '').fillna(0).astype('int64')
data['교통수단'] = data['교통수단'].str.replace(',', '').fillna(0).astype('int64')
data['유원지'] = data['유원지'].str.replace(',', '').fillna(0).astype('int64')
data['학교'] = data['학교'].str.replace(',', '').fillna(0).astype('int64')
data['금융기관'] = data['금융기관'].str.replace(',', '').fillna(0).astype('int64')
data['기타'] = data['기타'].str.replace(',', '').fillna(0).astype('int64')
```
