```python
import numpy as np
import pandas as pd
```

```python
data2019 = pd.read_table('./crime/2019.txt')
data2020 = pd.read_table('./crime/2020.txt')
```

```python
data2019 = data2019.drop(0)
data2020 = data2020.drop(0)
```

```python
data = pd.merge(data2019, data2020, how='outer')
```

```python
data
```

```python
data = data.drop(['합계.1', '살인.1', '강도.1', '강간강제추행.1', '절도.1', '폭력.1'], axis = 'columns')
```

```python
data['합계'] = data['합계'].str.replace(',', '').astype('int64')
data['강간강제추행'] = data['강간강제추행'].str.replace(',', '').astype('int64')
data['절도'] = data['절도'].str.replace(',', '').astype('int64')
data['폭력'] = data['폭력'].str.replace(',', '').astype('int64')
```

```python
data
```

```python
data['살인'] = data['살인'].replace('-', 0).astype('int64')
```

```python
data['강도'] = data['강도'].replace('-', 0).astype('int64')
```

```python
data.info()
```

```python
data.to_excel('./crime/crime_total(2019-2020).xlsx', index=False)
```
